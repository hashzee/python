```python
'''
Assignment No. 6 
================================================================
Question No. 1
--------------
Define Object Oriented Programming Language?

Answer: 
-------
As its name implplies, Object Oriented Programming is a programming style that is based on objects.
All the entities in this programming model are treated as objects. We define objects with their properties and functionalities.

Question No. 2
--------------
List down the Benefits of OOP?

Answer: 
-------
1) OOP offers great reusability 
2) It is quite extendable through inheritence 
3) Easier to maintain
4) It supports agile development
5) 

Question No. 3
--------------
Differentiate between function and method?

Answer: 
-------
Methods are special functions defined for an object inside its class.
Methods are called through object association while function are called by its name. In simple words if a function is 
defined inside a class (Object's class) it will be called a Method.

Question No. 4
--------------
Define the following terms:
1. Class
2. Object
3. Attribute
4. Behavior

Answer: 
-------
Class:
In OOP - a Class can be defined as a BLUEPRINT to create objects. It defines what an object will be composed of and how
that object will behave. For example, CAR can be a class. 

Object: 
In OOP,an Object is an implementation of a class - in other words, its an instance of a class. For example, in the 
definition of Class above, we say, CAR is a class. Now, think of Toyota Vitz, this is an Object of Class CAR.

Attribute: 
Each object has some fact or information about it. In OOP paradigm, attributes are structured in a class. Take the 
example of class CAR, the type of car (hatchback, sedan, suv), its color, engine capacity etc are good examples of its 
attributes. 

Behavior:
Its like what the instance of a class (an object) will do or perform or act based on a given instruction. For example, 
if we have a class of Mammal, we can make 2 objects Cat and Dog. Both can have a method 'Speak' however based on the 
message passed / or internal state change, the method will result in Meow or Bark. 

Question 5:
-----------
Write a code in python in which create a class named it Car which
have 5 attributes such like (model, color and name etc.) and 3
methods. And create 5 object instance from that class.

'''
class Car:
    def __init__(this, make, model, cc, transmission, carType):
        this.make = make
        this.model = model
        this.cc = cc
        this.transmission = transmission
        this.type = carType
    
    def showInfo(this):
        print("This is a "+this.make+' '+this.model+', a '+str(this.cc)+'cc '+this.transmission+' '+this.type)
    
    def calcMilage(this):
        cc = int(this.cc)
        if cc < 1000:
            print('Milage is 11-14km/l approx.')
        elif cc == 1000:
            print('Milage is 10km/l approx.')
        elif cc == 1300:
            print('Milage is 8-9km/l approx.')        
        elif cc == 1500:
            print('Milage is 7-8km/l approx.')
        elif cc == 2000:
            print('Milage is 5km/l approx.')
        elif cc > 2000:
            print('Milage is WORST.')
        else:
            print('Milage is unknown')
            
    def setCC(this,cc):
        this.cc = int(cc)

# END CLASS STRUCTURE
vitz = Car('Toyota','Vitz',1000,'Automatic','Hatchback')
vezel = Car('Honda','Vezel',1500,'Automatic','Mini SUV')
city = Car('Honda','City',1300,'Manual','Sedan')
thundra = Car('Toyota','Thundra',5500,'Manual','Truck')
bravo = Car('United','Brave',1000,'Manual','Hatchback')


vitz.showInfo()
vitz.calcMilage()
vitz.setCC(660)
vitz.showInfo()
vitz.calcMilage()
print("\n--------------\n")
vezel.showInfo()
vezel.calcMilage()
print("\n--------------\n")
city.showInfo()
city.calcMilage()
print("\n--------------\n")
thundra.showInfo()
thundra.calcMilage()
print("\n--------------\n")
bravo.showInfo()
bravo.calcMilage()

        
```
